```python
data = pd.read_excel('/content/drive/MyDrive/졸작/data/bigkinds_news data.xlsx')
```

```python
data
```

```python
keyword_list = []
for i in data['키워드']:
  keyword_list.append(i)
```

```python
keyword_list
```

```python
# 빈 리스트 생성
keyword_li_res = []

# 각 문자열을 분리하고 결과 리스트에 추가
for keywords in keyword_list:
    keyword_li_res.extend(keywords.split(','))

keyword_li_res
```

```python
len(keyword_li_res)
```

```python
# 공백제거
# keyword_li_res = keyword_li_res.strip()
keyword_li_res = [keyword.strip() for keyword in keyword_li_res]

keyword_li_res
```

```python
# 중복값 제거
keyword_li_res = list(set(keyword_li_res))
keyword_li_res
```

```python
len(keyword_li_res)  # 중복 제거 된듯? 4062910 --> 110427
```

```python
keyword_li_res = pd.DataFrame(keyword_li_res)
keyword_li_res
```
