```python
import pandas as pd
data = pd.read_excel('/content/drive/MyDrive/workspace/data/result.xlsx')
```

```python
data.head()
```

```python
df = pd.DataFrame(data)
df.head()
```

```python
# Split the '이해관계자' column and explode it
df = df.assign(이해관계자=df['이해관계자'].str.split(', ')).explode('이해관계자')

df.head()
```

```python
sh_code = pd.read_excel('/content/drive/MyDrive/workspace/data/대분류 code.xlsx')
mc_code = pd.read_excel('/content/drive/MyDrive/workspace/data/data/언론사 code.xlsx')
sh_list = pd.read_excel('/content/drive/MyDrive/workspace/data/data/(최종)이해관계자 총리스트.xlsx')
```

```python
sh_list.head()
sh_list.drop(['소분류'], axis=1, inplace=True)
sh_list.head()
```

```python
sh_code
```

```python
mc_code.head()
```

```python
merged_df = pd.merge(sh_list, sh_code, left_on='중분류', right_on='중분류 종류', how='left')
merged_df = merged_df.drop('중분류 종류', axis=1)
merged_df
```

```python
merged_df.drop(['대분류', '중분류'], inplace=True, axis=1)
merged_df.head()
```

```python
new_df = pd.merge(df, merged_df, left_on='이해관계자', right_on='이해관계자 리스트', how='left')

# Drop the '이해관계자 리스트' column as it is no longer needed
new_df = new_df.drop('이해관계자 리스트', axis=1)

# Rename the '중분류 코드' column to '이해관계자 코드'
new_df = new_df.rename(columns={'중분류 코드': '이해관계자 코드'})
```

```python
new_df.head()
```

```python
# Merge new_df with mc_code on '언론사' and '언론사 명칭'
total_df = pd.merge(new_df, mc_code, left_on='언론사', right_on='언론사 명칭', how='left')

# Drop the '언론사 명칭' column as it is no longer needed
total_df = total_df.drop('언론사 명칭', axis=1)
```
